### **The data were obtained in a survey of students math and portuguese language courses in secondary school. It contains a lot of interesting social, gender and study information about students. You can use it for some EDA or try to predict students final grade.**

Content:

Attributes for both student-mat.csv (Math course) and student-por.csv (Portuguese language course) datasets:

- 1. **school** - student's school (binary: 'GP' - Gabriel Pereira or 'MS' - Mousinho da Silveira)
- 2. **sex** - student's sex (binary: 'F' - female or 'M' - male)
- 3. **age** - student's age (numeric: from 15 to 22)
- 4. **address** - student's home address type (binary: 'U' - urban or 'R' - rural)
- 5. **famsize** - family size (binary: 'LE3' - less or equal to 3 or 'GT3' - greater than 3)
- 6. **Pstatus** - parent's cohabitation status (binary: 'T' - living together or 'A' - apart)
- 7. **Medu** - mother's education (numeric: 0 - none, 1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)
- 8. **Fedu** - father's education (numeric: 0 - none, 1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)
- 9. **Mjob** - mother's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')
- 10. **Fjob** - father's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')
- 11. **reason** - reason to choose this school (nominal: close to 'home', school 'reputation', 'course' preference or 'other')
- 12. **guardian** - student's guardian (nominal: 'mother', 'father' or 'other')
- 13. **traveltime** - home to school travel time (numeric: 1 - 1 hour)
- 14. **studytime** - weekly study time (numeric: 1 - 10 hours)
- 15. **failures** - number of past class failures (numeric: n if 1<=n<3, else 4)
- 16. **schoolsup** - extra educational support (binary: yes or no)
- 17. **famsup** - family educational support (binary: yes or no)
- 18. **paid** - extra paid classes within the course subject (Math or Portuguese) (binary: yes or no)
- 19. **activities** - extra-curricular activities (binary: yes or no)
- 20. **nursery** - attended nursery school (binary: yes or no)
- 21. **higher** - wants to take higher education (binary: yes or no)
- 22. **internet** - Internet access at home (binary: yes or no)
- 23. **romantic** - with a romantic relationship (binary: yes or no)
- 24. **famrel** - quality of family relationships (numeric: from 1 - very bad to 5 - excellent)
- 25. **freetime** - free time after school (numeric: from 1 - very low to 5 - very high)
- 26. **goout** - going out with friends (numeric: from 1 - very low to 5 - very high)
- 27. **Dalc** - workday alcohol consumption (numeric: from 1 - very low to 5 - very high)
- 28. **Walc** - weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)
- 29. **health** - current health status (numeric: from 1 - very bad to 5 - very good)
- 30. **absences** - number of school absences (numeric: from 0 to 93)

These grades are related with the course subject, Math or Portuguese:

- 31. **G1** - first period grade (numeric: from 0 to 20)
- 32. **G2** - second period grade (numeric: from 0 to 20)
- 33. **G3** - final grade (numeric: from 0 to 20, output target)

```python
def z_G3(dataframe):
    mu = dataframe['G3'].mean()
    sd = dataframe['G3'].std()
    z = (dataframe['G3']-mu)/sd
    dataframe['z'] = z
    return dataframe
math = z_G3(math)
por = z_G3(por)
```

```python
print(math[(math['z']<-2)|(math['z']>2)].shape)
math[(math['z']<-2)|(math['z']>2)]
```

```python
print(por[(por['z']<-2)|(por['z']>2)].shape)
por[(por['z']<-2)|(por['z']>2)]
```

- Now delete data for those have z(G3) less than -2.2 on both Math and Portuguese

```python
def drop_outlier(dataframe):
    idx = dataframe[dataframe['z']< -2.2].index
    dataframe.drop(idx,inplace=True)
    return dataframe
math = drop_outlier(math)
por = drop_outlier(por)
```

## 3.1 Function for LabelEncoding and StandardScaler

```python
def labelEncode(dataframe,columns):
    
    '''
    This function is for labeling nominal data. The function takes two arguments which are dataframe and nominal columns. 
    It returns:
    1) dataframe with labeled columns_ ,
    2) lb_object which would be used later during after prediction process, and 
    3) classes of those columns that are labeled
    '''
    
    from sklearn.preprocessing import LabelEncoder
    classes = dict()
    lb_object = dict()
    for column in columns:
        lb = LabelEncoder()
        after_lb = lb.fit_transform(dataframe[column])
        classes[column] = lb.classes_
        dataframe[column+'_lb'] = after_lb
        lb_object[column] = lb
    return dataframe, lb_object, classes

def standardScaler(dataframe,columns):
    '''
    This function is for z-standard normalization. This function takes two arguments which are dataframe and numeric columns.
    It returns:
    1) dataframe with standardized columns, 
    2) mean of data in such columns as dictionary datatype, 
    3) standard deviation of data in such columns as dictionary datatype and
    4) stantdard scaler object. This will be used later using prediction new data for converting data to be the same distribution as training data
    '''
    
    from sklearn.preprocessing import StandardScaler
    mean = dict()
    sd = dict()
    scaler = StandardScaler()
    scaler.fit(df[columns])
    for index,column in enumerate(columns):
        mean[column] = scaler.mean_[index]
        sd[column] = np.sqrt(scaler.var_)[index]
    dataframe[columns] = scaler.transform(dataframe[columns])    
    return dataframe, scaler, mean, sd, 
```

```python
def plot_model_performance(dataframe):
    '''
    This function is for plotting model performance (Measure accuracy)
    '''
    cols=dataframe.columns
    dataframe[cols].plot.box(figsize=(9,5))
    plt.title('Model Performance (Accuracy)')
    plt.xlabel('Model')
    plt.ylabel('Accuracy')
    for i,col in enumerate(cols):
        plt.text(0.5+i,dataframe[col].mean(),'mean: {:.2f}%'.format(dataframe[col].mean()*100))
    return plt.show()

def plot_confusion_matrix(y_test,y_pred):
    '''
    This function is for plotting confusion matrix to evaluate model performance
    '''
    cm = confusion_matrix(y_test,y_pred)
    target_class = np.sort(y_test.unique())
    plt.rcParams.update({'font.size':13})
    plt.figure(figsize=(8,6))
    ax = sns.heatmap(cm,annot=True,cmap='YlGnBu')
    ax.set(title='Confusion Matrix',xlabel='Prediction',ylabel='Actual')
    ax.xaxis.set_ticklabels(target_class)
    ax.yaxis.set_ticklabels(target_class)
    plt.show()
```

```python
df = df.sample(frac=1).reset_index(drop=True) # shuffle data
n_training_size = 300
train = df.iloc[:n_training_size,:]
test = df.iloc[n_training_size+1:,:]

X_test = test.drop(['G3'],axis=1)
y_test = test['G3']

target = ['G3']
features = df.drop(target,axis=1).columns

k_folds = 10
models = [DecisionTreeClassifier(), 
          RandomForestClassifier(),
          SVC(kernel='rbf'),
          GaussianNB()]   # Model instantiation
model_names = ['DecisionTree','RandomForest','SupportVectorMachine','NaiveBayes']
model_acc_score = dict()

for i,model in enumerate(models):
    train = df.iloc[:n_training_size,:]
    acc_score = list()
    for k in range(k_folds):
        sample_val = train.shape[0]//k_folds
        validation = train[sample_val*k:(k+1)*sample_val]
        # x_val and y_val
        X_val = validation[features]
        y_val = validation[target]
        valid_index = validation.index

        # train set
        train_index = [i for i in range(0,train.shape[0]) if i not in valid_index]
        train_Xy = train.iloc[train_index]
        X_train = train_Xy[features]
        y_train = train_Xy[target]
        
        # fit and predict model
        model.fit(X_train,y_train)
        y_pred = model.predict(X_val)
        acc_score.append(accuracy_score(y_val,y_pred))
    model_acc_score[model_names[i]] = acc_score
df_acc_score = pd.DataFrame(model_acc_score,index=['Round'+str(i+1) for i in range(k_folds)])
print('Mean Accuracy - Summary of {} Folds Cross Validation'.format(k_folds))
print('_______________________')
print(df_acc_score.mean())
print(df_acc_score)
plot_model_performance(df_acc_score)
```

The mean accuracy score offered by each model are very low, approximately 10%.Nonetheless, if we subtly consider the predicted value and true value, we can clearly see that predicted values only slightly differ from true value. Therefore, it is reasonable to create a new metric that loosely measure the prediction accuracy. To elaborate, this metric is going to give the score to a model that offers predicted value differed from true value by 2 unit as correct prediction. For example, true value = 7, predicted value = 9 -> correct prediction

```python
def prediction_adjusted(y_test,y_pred,diff=2):
    '''
    This function is for adjusting the value of the prediction. 
    If the model predict the value differed from true value by 1 unit, 
    it is going to change such predicted value to be equal to true value.
    '''
    y_test = y_test.values.ravel()
    y_pred = np.array(y_pred)
    y_pred_adj = np.where(np.absolute(y_pred-y_test)<=diff,y_test,y_pred)
    return y_pred_adj
```

```python
# Hyperparameters
estimators = [100,300,500,800,1200]
max_features = ['auto','log2']
max_depth = [5, 8, 15, 25, 30]
min_samples_split = [2, 5, 10, 15, 20]
min_samples_leaf = [1, 2, 5, 10]
param = {'n_estimators': estimators,
         'max_features':max_features,
         'max_depth':max_depth,
         'min_samples_split':min_samples_split,
         'min_samples_leaf':min_samples_leaf}
params_dict = dict()
```

```python
df_por = df_por.sample(frac=1).reset_index(drop=True) # shuffle data
n_training_size = 500
train = df_por.iloc[:n_training_size,:]
test = df_por.iloc[n_training_size+1:,:]

X_test = test.drop(['G3'],axis=1)
y_test = test['G3']

target = ['G3']
features = df.drop(target,axis=1).columns

k_folds = 10
models = [DecisionTreeClassifier(), 
          RandomForestClassifier(),
          SVC(kernel='rbf'),
          GaussianNB()]   # Model instantiation
model_names = ['DecisionTree','RandomForest','SupportVectorMachine','NaiveBayes']
model_acc_score = dict()

for i,model in enumerate(models):
    train = df_por.iloc[:n_training_size,:]
    acc_score = list()
    
    acc_score_train = list()
    
    for k in range(k_folds):
        sample_val = train.shape[0]//k_folds
        validation = train[sample_val*k:(k+1)*sample_val]
        # x_val and y_val
        X_val = validation[features]
        y_val = validation[target]
        valid_index = validation.index

        # train set
        train_index = [i for i in range(0,train.shape[0]) if i not in valid_index]
        train_Xy = train.iloc[train_index]
        X_train = train_Xy[features]
        y_train = train_Xy[target]
        
        # fit and predict model
        model.fit(X_train,y_train)
        y_pred = model.predict(X_val)
        y_pred_adj = prediction_adjusted(y_val,y_pred)
        acc_score.append(accuracy_score(y_val,y_pred_adj))
       
    model_acc_score[model_names[i]] = acc_score

df_por_acc_score = pd.DataFrame(model_acc_score,index=['Round'+str(i+1) for i in range(k_folds)])
print('Mean Accuracy - Summary of {} Folds Cross Validation'.format(k_folds))
print('_______________________')
print(df_por_acc_score.mean())
print(df_por_acc_score)

plot_model_performance(df_por_acc_score)
```
